#### 정규성 검정
- 정규분포를 따르는지 검정
- 가정: 독립성, 등분산성
- 양적자료 < 2000 이면 "샤피로-월크 테스트"
- 양적자료 >= 2000 이면 "앤더슨-달링 테스트" 

```python
# Shapiro-Wilk Test
from scipy.stats import shapiro
print(shapiro(red_quality))
print(shapiro(white_wine_quality))

## red_quality 와 white_quality가 유의수준 0.05에서 귀무가설을 기각한다.
## 따라서 정규성을 따른다고 할 수 없다.
```

```python
# D'Agostino's K^2 Test - 왜도와 첨도가 정규분포와 일치하는 지를 판단
from scipy.stats import normaltest
print(normaltest(red_quality))
print(normaltest(white_quality))
```

```python
# Kolmogorove-Smirnov test
from scipy.stats import kstest

print(kstest(red_quality, 'norm'))
print(kstest(white_quality, 'norm'))

## red_quality 와 white_quality가 유의수준 0.05에서 귀무가설을 기각한다.
## 따라서 정규성을 따른다고 할 수 없다.
```

#### 독립성 검정
- 데이터셋이 독립성을 따르는지 검정

```python

# Chi-squared test of independence 카이제곱 검정
from scipy.stats import chi2_contingency
df_crosstab = pd.crosstab(red_quality,white_quality)
chi2_contingency(df_crosstab)


## 순서대로 검정통계량, p-value, 자유도, 기대도수
## red_quality 와 white_quality가 유의수준 0.05에서 귀무가설을 기각한다.
## 따라서 red_quality 와 white_quality는 서로 관계가 있다고 볼 수 있다.
```

#### 등분산 검정
- 데이터셋의 분산이 같은지 검정

```python
# Levene's test (정규분포와 관계없이 연속적인 분포로부터 데이터를 구했을 때 사용한다.)
from scipy.stats import levene
stats.levene(red_quality, white_quality)

## red_quality와 white_quality가 유의수준 0.05에서 귀무가설을 기각하지 못한다.
## 따라서 분산이 같다. 라고 할 수 있다.
```

```python
# Bartlett's test(비정규성에 민감하므로 정규성을 따를 때만 사용한다.)
from scipy.stats import bartlett
bartlett(red_quality,white_quality)

## red_quality와 white_quality가 유의수준 0.05에서 귀무가설을 기각한다.
## 따라서 분산이 같지않다. 라고 할 수 있다.
```

```python
# stats.ttest_ind() 함수를 사용하여 t-검정을 하고 두 그룹 간 차이를 확인
stats.ttest_ind(red_quality, white_quality ,equal_var = True)
```
