```python
Hy_geoData = pd.read_csv('hollysgeo.shp.csv',encoding='cp949', engine='python')
Hy_geoData
```

```python
Hy_geoData.head()
```

```python
map_Hy = folium.Map(location = [36.54100095967711, 128.79629698726797], zoom_start = 15)
map_Hy
```

```python
for i, store in Hy_geoData.iterrows(): 
    folium.Marker(location=[store['위도'],store['경도']],popup=store['store'],
                  icon = folium.Icon(color='red',icon='star')).add_to(map_Hy)
```

```python
map_Hy.save('map_Hy.html')
```
